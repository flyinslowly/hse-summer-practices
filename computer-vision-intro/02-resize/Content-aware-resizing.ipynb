```python
import os
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline

from skimage.io import imread
from skimage.color import convert_colorspace
from skimage.filters import prewitt
from scipy.ndimage.interpolation import rotate
from time import time
```

```python
def count_energy_matrix(image_brightness, mask=None):
    image_energy = np.ones(image_brightness.shape)
    height, width = image_brightness.shape
    for x in range(height):
        for y in range(width):
            if (y == 0):
                dy = image_brightness[x][y + 1] - image_brightness[x][y]
            elif (y == width - 1):
                dy = image_brightness[x][y] - image_brightness[x][y - 1]
            else:
                dy = image_brightness[x][y + 1] - image_brightness[x][y - 1]
            
            if (x == 0):
                dx = image_brightness[x + 1][y] - image_brightness[x][y]
            elif (x == height - 1):
                dx = image_brightness[x][y] - image_brightness[x - 1][y]
            else:
                dx = image_brightness[x + 1][y] - image_brightness[x - 1][y]
                
            if (mask is None):
                image_energy[x][y] = np.sqrt(dx ** 2 + dy ** 2)
            else:
                if (mask[x][y] == 0):
                    image_energy[x][y] = np.sqrt(dx ** 2 + dy ** 2)
                elif (mask[x][y] == 1):
                    image_energy[x][y] = height * width * 256
                elif (mask[x][y] == -1):
                    image_energy[x][y] = (-1) * height * width * 256
                    
    return image_energy
```

```python
def get_minimums_id(array):
    minimum = -1
    id_min = -1
    for i in range(array.shape[0]):
        if id_min == -1 or minimum > array[i]:
            id_min = i
            minimum = array[i]
    return id_min

def get_min_seam(image_energy):
    seam_matrix = np.zeros(image_energy.shape)
    seam_matrix[0] = image_energy[0]
    for x in range(1, seam_matrix.shape[0]):
        for y in range(seam_matrix.shape[1]):
            seam_matrix[x][y] += image_energy[x][y]
            if y == 0:
                seam_matrix[x][y] += min(seam_matrix[x - 1][y], seam_matrix[x - 1][y + 1])
            elif y == seam_matrix.shape[1] - 1:
                seam_matrix[x][y] += min(seam_matrix[x - 1][y - 1], seam_matrix[x - 1][y])
            else:
                seam_matrix[x][y] += min(seam_matrix[x - 1][y - 1], seam_matrix[x - 1][y], seam_matrix[x - 1][y + 1])
    cur_y = get_minimums_id(seam_matrix[seam_matrix.shape[0] - 1])
    seam_ind = [(seam_matrix.shape[0] - 1, cur_y)]
    while len(seam_ind) < seam_matrix.shape[0]:
        cur_height = seam_matrix.shape[0] - len(seam_ind)
        cur_height -= 1
        cur_y = cur_y - 1 + get_minimums_id(seam_matrix[cur_height][cur_y - 1:cur_y + 2])
        seam_ind.append((cur_height, cur_y))
    return tuple(reversed(seam_ind))
#print(get_min_seam(image_energy))
```

```python
def resize(image, seam, mode, mask):
    carve_mask = np.zeros(image.shape[:2], dtype='uint8')
    if mode == 'shrink':
        height, width, depth = image.shape
        resized_image = np.zeros((height, width - 1, depth), dtype='uint8')
        resized_mask = np.zeros((height, width - 1), dtype='int8')
        for x in range(height):
            resized_image[x] = np.delete(image[x], seam[x][1], axis=0)
            resized_mask[x] = np.delete(mask[x], seam[x][1])
            carve_mask[seam[x]] = 1

    elif mode == 'expand':
        heigth, width, depth = image.shape
        resized_image = np.zeros((height, width + 1, depth), dtype='uint8')
        resized_mask = np.zeros((height, width + 1), dtype='int8')
        for x in range(height):
            if seam[x][1] == img.shape[1] - 1:
                new_pixel = (img[seam[x]] / 2 + img[seam[x][0], seam[x][1] - 1] / 2)
            else:
                new_pixel = (img[seam[x]] / 2 + img[seam[x][0], seam[x][1] + 1] / 2)
            carve_mask[seam[x]] = 1
            resized_image[x] = np.insert(image[x], seam[x][1], new_pixel, axis=0)
            resized_mask[x] = np.insert(mask[x], seam[x][1], 0)
            resized_mask[seam[x]] = 1
    return resized_image, carve_mask, resized_mask
```

```python
image = imread('img.png')
mask = imread('mask.png')

mask = ((mask[:,:,0]!=0)*(-1) + (mask[:,:,1]!=0)).astype('int8')
# 2D mask + {-1, 0, 1} + RGB image
plt.imshow(mask)

def get_brightness(img):
    return convert_colorspace(img, 'RGB', 'YUV')[:,:,0]

image_brightness = get_brightness(image)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(30, 30))
ax1.imshow(image)
ax2.imshow(image_brightness, cmap='gray')
image_energy = count_energy_matrix(image_brightness)
```

```python
%%time

seam = get_min_seam(image_energy)

#fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15, 15))
#ax1.imshow(image)

for i in range(1):
    resized_img, carve_mask, resized_mask = resize(image, seam, 'shrink', mask)
    image = resized_img
    brightness = get_brightness(image)
    image_energy = count_energy_matrix(brightness)
    seam = get_min_seam(image_energy)
    mask = resized_mask

#ax2.imshow(image, cmap='gray')
print(seam)
```
